```python
#DSA
class Stack:
    def __init__(self):
        self.num = []
    def push(self,number):
        self.num.append(number)
    def pop(self):
        return self.num.pop()
    def peek(self):
        return self.num[-1]
    def is_empty(self):
        if self.num == [] :
            return True
        else:
            return False
    def display(self):
        print(self.num)
        
    
```

```python
s = Stack()
choice = 1
while(choice):
    print("\nFor Push Element Input 1")
    print("For Pop Element Input 2")
    print("For Peek Element Input 3")
    print("For Check Is_empty Stack Input 4")
    print("For Display Input 5")
    print("For Quit Input 0")
    choice = int(input("Enter Choice="))
    if choice == 1:
        number= int(input("Enter Number You Want to Push"))
        s.push(number)
    if choice == 2:
        pop=s.pop()
        print(pop)
    if choice == 3:
        peek=s.peek()
        print("Peek Element is=",peek)
    if choice == 4:
        is_empty=s.is_empty()
        print("\n",is_empty)
    if choice == 5:
        s.display()
    if choice == 0:
        break;
    
        
```

```python
class Queue:
    def __init__(self):
        self.queue = []
    def enqueue(self,num):
        self.queue.append(num)
    def dequeue(self):
        return self.queue.pop(0)
    def peek(self):
        return self.queue[0]
    def is_empty(self):
        if self.queue == [] :
            return True
        else:
            return False
    def display(self):
        print(self.queue)
        
```

```python
q = Queue()
choice = 1
while(choice):
    print("\nFor Enqueue Element Input 1")
    print("For dequeue Element Input 2")
    print("For Peek Element Input 3")
    print("For Check Is_empty Stack Input 4")
    print("For Display Input 5")
    print("For Quit Input 0")
    choice = int(input("Enter Choice="))
    if choice == 1:
        number= int(input("Enter Number You Want to enqueue "))
        q.enqueue(number)
    if choice == 2:
        dequeue=q.dequeue()
        print("Dequeue Element is",dequeue)
        
    if choice == 3:
        peek=q.peek()
        print("Peek Element is=",peek)
    if choice == 4:
        is_empty=q.is_empty()
        print("\n",is_empty)
    if choice == 5:
        q.display()
    if choice == 0:
        break;
```

```python
#Python Coding
# Function for nth Fibonacci number
 
def Fibonacci(n):
    if n<= 0:
        print("Incorrect input")
    # First Fibonacci number is 0
    elif n == 1:
        return 0
    # Second Fibonacci number is 1
    elif n == 2:
        return 1
    else:
        return Fibonacci(n-1)+Fibonacci(n-2)
 
print(Fibonacci(10))
 
```

```python
class BinarySearchTree:
    def __init__(self,key):
        self.key = key
        self.lchild = None
        self.rchild = None
    #For Inserting
    def insert(self,data):
        if self.key is None:
            self.key = data
        if self.key == data:
            return
        if self.key > data:
            if self.lchild:
                self.lchild.insert(data)
            else:
                self.lchild = BinarySearchTree(data)
        else:
            if self.rchild:
                self.rchild.insert(data)
            else:
                self.rchild = BinarySearchTree(data)
    #For Searching Node Is Present Or not
    def search(self,data):
        if  self.key == data:
            print("Node Is Found")
            return True
        if data< self.key:
            if self.lchild:
                self.lchild.search(data)
            else:
                print("Node Is Not Present")
        else:
            if self.rchild:
                self.rchild.search(data)
            else:
                print("Node Is Not Present")
    #for Deletion node
    def delete(self,data):
        if self.key is None:
            print("Tree Is Empty")
            return
        if data<self.key:
            if self.lchild:
                self.lchild = self.lchild.delete(data)
            else:
                print("Given node Is not Present")
        elif data > self.key:
            if self.lchild:
                self.rchild = self.rchild.delete(data)
            else:
                print("Given node Is not Present")
        #if node have another Child node then delete the node and make it parent the child trthe one
        else:
            if self.lchild is None:
                temp = self.rchild
                self = None
                return temp
            if self.rchild is None:
                temp = self.lchild
                self = None
                return temp
            node = self.rchild #it will find until smallest node is not found
            while node.lchild:
                node = node.lchild 
            self.key = node.key
            self.rchild = self.rchild.delete(node.key)
        return self
       #this will return deleted node 
#for size        
def size(node):
        if node is None:
            return 0
        else:
            return (size(node.lchild)+1+size(node.rchild))
        
                
            
        
        
```

```python
root = BinarySearchTree(10)
list1 = [20,4,30,4,1,5,6]
for i in list1:
    root.insert(i)
root.search(4)
root.search(65)
size(root)
```
